# <center>Attack on the Clone Lab<center> <a id="chapter-12"></a>

This is a tutorial demonstrating how to set up a foundation network of virtual machines (VM) on Ubuntu 16.04 which can be used to practice digital forensics, networking and penetration testing. The foundation network contains 3 routers and 3 pcs, a metasploitable 2 LAMP server and a Kali Linux host. Other servers and hosts, with varying operating systems, etc. can be added to the network to customize it.

## Set up Quagga router <a id="chapter-7"></a>


**Configure the router daemon file**
```
sudo nano /etc/quagga/daemons
```

Change the settings in the end of file to enable zebra and ospfd daemons to look like: 
```
    zebra=yes
    bgpd=no
    ospfd=yes
    ospf6d=no
    ripd=no
    ripngd=no
    isisd=no
    babeld=no
```
**Copy the configuration files for zebra and ospfd daemons from /usr/share/doc/quagga/examples into the /etc/quagga configuration directory.**

```
sudo cp /usr/share/doc/quagga/examples/ospfd.conf.sample /etc/quagga/ospfd.conf
sudo cp /usr/share/doc/quagga/examples/zebra.conf.sample /etc/quagga/zebra.conf
```

**Change ownership and permissions of daemon zebra.conf and ospfd.conf files, so quagga and quaggavty groups are owners, and there is rw permissions for both.**

```
sudo chown quagga.quaggavty /etc/quagga/daemons /etc/quagga/zebra.conf /etc/quagga/ospfd.conf
sudo chmod 660 /etc/quagga/daemons /etc/quagga/zebra.conf /etc/quagga/ospfd.conf
```

**Add users who will configure the router to the quagga and quaggavty groups.**
```
sudo nano /etc/group
```

At the bottom of the file it should look like something like:
```
quagga:x:1003:[user]
quaggavty:x:1004:[user]
```
**Start the routing suite**
```
sudo /etc/init.d/quagga start
```

**Set up environment to avoid vtysh end problem.**
```
sudo nano /etc/bash.bashrc
```
Copy this into the bottom line of the file:
```
export VTYSH_PAGER=more
VTYSH_PAGER=more
sudo nano /etc/environment
```

Copy this into the last line of the file:
```
VTYSH_PAGER=more

```
**Reboot the entire machine**
```
shutdown -r 0
```
**Log back in and start vtysh, which is the integrated shell for the Quagga routing engine that will allow you to configure the ip routing tables of the router.**
```
vtysh
```
**If you receive a message that says something like: "Failed to connect to any daemons"
Verify that quagga and quaggavty are the respective owner and group of the files in /etc/quagga**
```
ls -la /etc/quagga/
```
**Make sure that quagga service is active (running)**
```
sudo service quagga status 
```
**Also make sure that you restarted the entire system.**

## Configure the routers <a id="chapter-8"></a>


We are going to configure the 3 routers according to Brian Linkletter's topology.



**Enable port forwarding through VirtualBox, to allow ssh into each router from your host machine.**

In VirtualBox:
    
    Right click on the vm in the left banner and choose "Settings... Ctrl+S"
    
    Choose "Network" from the menu in the left hand banner.
    
    Click on "Adapter 1" 
    
    Verify the adapter says, "Attach to: NAT"
    
    Under "Advanced," click on "port forwarding."
    
    For router 1:
        protocol: ssh
        Host Port: 14601
        Guest Port: 22

    Repeat this process for router2 and router3, changing the host port to 14602 and 14603, respectively. 
        
**Headless start all three routers**

In VirtualBox:
    
    Right click on "Start" and choose "Headless Start"
    
**ssh into routers from host machine**

In the command line, use the formula
   ```
   ssh -l [user] -p [Host Port] localhost
   ```
It should look something like: 
   ```
   ssh -l jane -p 14601 localhost
   ``` 

**Configure Router 1**

Type the following commands once you are in the vtysh shell:
   ```
   configure terminal
   hostname router1
    router ospf
     network 192.168.1.0/24 area 0
     network 192.168.100.0/24 area 0 
     network 192.168.101.0/24 area 0 
     passive-interface enp0s8    
     exit
    interface enp0s8
     ip address 192.168.1.1/24
     exit
    interface enp0s9
     ip address 192.168.100.1/24
     exit
    interface enp0s10
     ip address 192.168.101.1/24
     exit
    ip forwarding
    exit
   write
   exit
   ```

Make additional rules in ufw to allow network traffic

```
sudo ufw allow from 192.168.1.0/24
sudo ufw allow from 192.168.100.0/24
sudo ufw allow from 192.168.101.0/24
```


**Configure Router 2**

Type the following commands once you are in the vtysh shell:
   ```
   configure terminal
   hostname router2
    router ospf
     network 192.168.1.0/24 area 0.0.0.0
     network 192.168.100.0/24 area 0.0.0.0
     network 192.168.102.0/24 area 0.0.0.0
     passive-interface enp0s8    
     exit
    interface enp0s8
     ip address 192.168.2.1/24
     exit
    interface enp0s9
     ip address 192.168.100.1/24
     exit
    interface enp0s10
     ip address 192.168.102.1/24
     exit
    ip forwarding
    exit
   write
   exit
  ```
Make additional rules in ufw to allow network traffic
```
sudo ufw allow from 192.168.2.0/24
sudo ufw allow from 192.168.100.0/24
sudo ufw allow from 192.168.102.0/24
```


**Configure Router 3**

Type the following commands once you are in the vtysh shell:
   ```
    configure terminal
    hostname router2
     router ospf
      network 192.168.3.0/24 area 0.0.0.0
      network 192.168.101.0/24 area 0.0.0.0
      network 192.168.102.0/24 area 0.0.0.0
      passive-interface enp0s8    
      exit
     interface enp0s8
      ip address 192.168.3.1/24
      exit
     interface enp0s9
      ip address 192.168.101.2/24
      exit
     interface enp0s10
      ip address 192.168.102.2/24
      exit
     ip forwarding
     exit
    write
    exit
   ```

Make additional rules in ufw to allow network traffic
```
sudo ufw allow from 192.168.3.0/24
sudo ufw allow from 192.168.101.0/24
sudo ufw allow from 192.168.102.0/24
```

**Note: If you type the wrong ip address in an interface here are the commands to delete it, starting from the vtysh command:**
```
vtysh
configure terminal
 interface [interface]
 no ip address [ip address]
 exit
exit
write
```

## Configure the PC's onto the network <a id="chapter-10"></a>


***Enable port forwarding through VirtualBox, to allow ssh into each pc from your host machine**

In VirtualBox:
    
    Right click on the vm in the left banner and choose "Settings... Ctrl+S"
    
    Choose "Network" from the menu in the left hand banner.
    
    Click on "Adapter 1" 
    
    Verify the adapter says, "Attach to: NAT"
    
    Under "Advanced," click on "port forwarding."
    
    For router 1:
        protocol: ssh
        Host Port: 14501
        Guest Port: 22

Repeat this process for router2 and router3, changing the host port to 14502 and 14503, respectively. 
        
**Headless start all three pc**

In VirtualBox:

    Right click on "Start" and choose "Headless Start"
    
**ssh into routers from host machine**

In the command line, use the formula:
```
ssh -l [user] -p [Host Port] localhost
```
It should look something like: 
```
ssh -l tom -p 14501 localhost
```
**log on**

**Change the host to pc1,2 or 3**
```
sudo nano /etc/hosts
```
**Change the hostname to pc1,2, or 3.**

```
sudo nano /etc/hostname
```
**Restart the computer so the changes are reflected.**
```
shutdown -r 0
```
**Log back onto the computer.**

**Add the interface configuration to the network interfaces file and set up a static route.**

pc1
```
auto enp0s8
iface enp0s8 inet static
   address 192.168.1.1
   netmask 255.255.255.0
up route add -net 192.168.0.0/16 gw 192.168.1.254 dev enp0s8
```
pc2
```
auto enp0s8
iface enp0s8 inet static
   address 192.168.2.1
   netmask 255.255.255.0
up route add -net 192.168.0.0/16 gw 192.168.2.254 dev enp0s8
```
pc3
```
auto enp0s8
iface enp0s8 inet static
   address 192.168.3.1
   netmask 255.255.255.0
up route add -net 192.168.0.0/16 gw 192.168.3.254 dev enp0s8
```

Multiple operating systems could be added to the networks, as well. Windows 7, 8 and 10, as well as Kali Linux and metasploitable could be integrated. Once customized, the lab should offer endless hours of hacker enjoyment. So go forth... have fun... and let me know what you accomplish!
